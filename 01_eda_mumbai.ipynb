```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

print("Loading Mumbai Housing Data from local file...")

# Use the correct path
file_path = 'data/mumbai_housing_data.csv'

# Check if file exists
if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("✓ Mumbai Dataset loaded successfully!")
else:
    print(f"❌ File not found at: {file_path}")
    print("Available files:", os.listdir('data'))

print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Check the actual columns in your dataset
print("Number of columns:", len(df.columns))
print("\nAll column names:")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

```python
# Get basic information about dataset
print("="*60)
print("DATASET INFORMATION")
print("="*60)

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nDataset shape:")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\nMissing values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
print(df.describe())
```

```python
# Print all column names to find the exact price column name
print("All columns in the dataset:")
print("="*60)
for i, col in enumerate(df.columns, 1):
    print(f"{i}. '{col}'")

print("\n" + "="*60)
print("Looking for price-related columns:")
price_cols = [col for col in df.columns if 'price' in col.lower()]
print(price_cols)

print("\n" + "="*60)
print("Looking for area-related columns:")
area_cols = [col for col in df.columns if 'area' in col.lower()]
print(area_cols)
```

```python
# Clean column names - remove leading/trailing spaces
df.columns = df.columns.str.strip()

print("Cleaned column names:")
print(df.columns.tolist())

print("\nNow trying to access Price column:")
print(f"✓ Price column found: {df['Price'].describe()}")
```

```python
# Create distribution plots for main numeric features
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Mumbai Housing Data - Feature Distributions', fontsize=16, fontweight='bold')

# Plot 1: Price distribution
axes[0, 0].hist(df['Price'], bins=40, edgecolor='black', color='#FF6B6B')
axes[0, 0].set_title('Price Distribution', fontweight='bold', fontsize=12)
axes[0, 0].set_xlabel('Price (₹)')
axes[0, 0].set_ylabel('Number of Properties')
axes[0, 0].grid(alpha=0.3)

# Plot 2: Area distribution
axes[0, 1].hist(df['Area'], bins=40, edgecolor='black', color='#4ECDC4')
axes[0, 1].set_title('Area Distribution', fontweight='bold', fontsize=12)
axes[0, 1].set_xlabel('Area (sqft)')
axes[0, 1].set_ylabel('Number of Properties')
axes[0, 1].grid(alpha=0.3)

# Plot 3: Bedrooms distribution
axes[0, 2].hist(df['No. of Bedrooms'], bins=10, edgecolor='black', color='#45B7D1')
axes[0, 2].set_title('Bedrooms Distribution', fontweight='bold', fontsize=12)
axes[0, 2].set_xlabel('Number of Bedrooms')
axes[0, 2].set_ylabel('Number of Properties')
axes[0, 2].grid(alpha=0.3)

# Plot 4: Price vs Area (scatter plot)
axes[1, 0].scatter(df['Area'], df['Price'], alpha=0.5, s=30, color='#FF7F50')
axes[1, 0].set_title('Price vs Area', fontweight='bold', fontsize=12)
axes[1, 0].set_xlabel('Area (sqft)')
axes[1, 0].set_ylabel('Price (₹)')
axes[1, 0].grid(alpha=0.3)

# Plot 5: Top 10 Locations
top_locations = df['Location'].value_counts().head(10)
axes[1, 1].barh(range(len(top_locations)), top_locations.values, color='#98D8C8')
axes[1, 1].set_yticks(range(len(top_locations)))
axes[1, 1].set_yticklabels(top_locations.index, fontsize=9)
axes[1, 1].set_title('Top 10 Locations', fontweight='bold', fontsize=12)
axes[1, 1].set_xlabel('Number of Properties')
axes[1, 1].grid(alpha=0.3, axis='x')

# Plot 6: Resale vs New properties
resale_counts = df['Resale'].value_counts()
axes[1, 2].bar(['New', 'Resale'], resale_counts.values, color=['#FFD700', '#FF6B6B'])
axes[1, 2].set_title('New vs Resale Properties', fontweight='bold', fontsize=12)
axes[1, 2].set_ylabel('Count')
axes[1, 2].grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('plots/01_mumbai_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Distribution plots saved to: plots/01_mumbai_distributions.png")
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

print(f"Number of numeric columns: {len(numeric_df.columns)}")
print("Numeric columns:")
print(numeric_df.columns.tolist())

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Create heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, 
            annot_kws={'size': 8})
plt.title('Correlation Matrix - Mumbai Housing Data', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('plots/02_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Correlation matrix saved!")

# Show correlation with Price
print("\n" + "="*60)
print("FEATURES CORRELATED WITH PRICE (Most Important First)")
print("="*60)

price_corr = numeric_df.corr()['Price'].sort_values(ascending=False)
print(price_corr)
```

```python
# Analyze top locations
print("="*60)
print("TOP 15 MOST EXPENSIVE LOCATIONS IN MUMBAI")
print("="*60)

# Group by location and calculate stats
locality_stats = df.groupby('Location').agg({
    'Price': ['mean', 'median', 'min', 'max', 'count'],
    'Area': 'mean',
    'No. of Bedrooms': 'mean'
}).round(2)

# Flatten the column names
locality_stats.columns = ['Avg Price', 'Median Price', 'Min Price', 'Max Price', 'Count', 'Avg Area', 'Avg Bedrooms']

# Sort by average price (descending)
locality_stats = locality_stats.sort_values('Avg Price', ascending=False)

print("\nTop 15 Most Expensive Locations:")
print(locality_stats.head(15))

# Visualization
fig, ax = plt.subplots(figsize=(12, 8))
top_15 = locality_stats.head(15)
colors = ['#FF6B6B' if i == 0 else '#4ECDC4' for i in range(len(top_15))]
ax.barh(range(len(top_15)), top_15['Avg Price'], color=colors)
ax.set_yticks(range(len(top_15)))
ax.set_yticklabels(top_15.index, fontsize=10)
ax.set_xlabel('Average Price (₹)', fontsize=11)
ax.set_title('Top 15 Most Expensive Locations in Mumbai', fontweight='bold', fontsize=13)
ax.grid(alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('plots/03_top_localities.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Location analysis saved!")
```

```python
# Analyze price ranges
print("="*60)
print("PRICE RANGE ANALYSIS")
print("="*60)

print(f"\n💰 Price Statistics (in ₹):")
print(f"   Minimum: ₹{df['Price'].min():,.0f}")
print(f"   Maximum: ₹{df['Price'].max():,.0f}")
print(f"   Average: ₹{df['Price'].mean():,.0f}")
print(f"   Median: ₹{df['Price'].median():,.0f}")
print(f"   Standard Deviation: ₹{df['Price'].std():,.0f}")

# Create price bins for distribution (in lakhs for Indian context)
price_in_lakhs = df['Price'] / 100000  # Convert to lakhs

price_bins = [0, 20, 50, 75, 100, 150, 250, 1000]
price_labels = ['<20L', '20-50L', '50-75L', '75-100L', '100-150L', '150-250L', '>250L']

df_temp = df.copy()
df_temp['Price Range'] = pd.cut(price_in_lakhs, bins=price_bins, labels=price_labels)

print(f"\n📊 Properties by Price Range (in Lakhs):")
price_range_dist = df_temp['Price Range'].value_counts().sort_index()
print(price_range_dist)

# Visualization
fig, ax = plt.subplots(figsize=(12, 6))
colors_price = ['#FF6B6B', '#FF8E8E', '#FFA07A', '#FFB6A3', '#FFD9B3', '#FFEEDD', '#98D8C8']
ax.bar(range(len(price_range_dist)), price_range_dist.values, color=colors_price)
ax.set_xticks(range(len(price_range_dist)))
ax.set_xticklabels(price_range_dist.index, fontsize=11)
ax.set_xlabel('Price Range (₹ Lakhs)', fontsize=12)
ax.set_ylabel('Number of Properties', fontsize=12)
ax.set_title('Distribution of Properties by Price Range', fontweight='bold', fontsize=13)
ax.grid(alpha=0.3, axis='y')

# Add count labels on bars
for i, v in enumerate(price_range_dist.values):
    ax.text(i, v + 20, str(v), ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('plots/04_price_ranges.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Price range analysis saved!")
```
